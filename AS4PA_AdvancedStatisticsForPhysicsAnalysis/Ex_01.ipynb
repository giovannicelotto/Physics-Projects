## Exercise 1 - vectors and data frames
The following table gives the volume, area, length and maximum and mean depths of some Scottish
lakes[1]. Create vectors, holding the lake’s name and all the parameters and build a dataframe
called scottish.lakes from the vectors
* 1 evaluate the highest and lowest volume and area lake
* 2 order the frame with respect to the area and determine the two largest area lakes
* 3 by summing up the areas occpupied by the lakes, determine the area of Scotland covered by water

```r
#define vectors
loch       <- c('Loch Ness', 'Loch Lomond', 'Loch Morar', 'Loch Tay','Loch Awe', 'Loch Maree', 'Loch Ericht', 
                'Locht Lochy', 'Loch Rannoch', 'Loch Shiel', 'Loch Katrine', 'Loch Arkaig','Loch Shin')

volume     <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
area       <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
length     <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
max_depth  <- c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49)
mean_depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)

#define the dataframe
scottish.lakes <- data.frame(row.names=loch, volume, area, length, max_depth, mean_depth)
scottish.lakes

#evaluate highest and lowest volume and area lakes
cat("The highest area is \t",    max(scottish.lakes['area']),  "\tkm^2 it belongs to \t",rownames(scottish.lakes)[which.max(area)],  "\n")
cat("The lowest area is \t",     min(scottish.lakes['area']),  "\tkm^2 it belongs to \t", rownames(scottish.lakes)[which.min(area)],  "\n")
cat("The highest volume is \t", max(scottish.lakes['volume']),"\tkm^3 it belongs to \t",rownames(scottish.lakes)[which.max(volume)],"\n")
cat("The lowest volume is \t",  min(scottish.lakes['volume']),"\tkm^3 it belongs to \t", rownames(scottish.lakes)[which.min(volume)],"\n")
```

2) order the frame with respect to the area and determine the two largest area lakes

```r
#order data frame with respect to the area
ordered_area <- scottish.lakes[order(scottish.lakes$area, decreasing = TRUE),]
ordered_area

#first two
cat("The two largest area lakes are \n1)\t",rownames(ordered_area)[1],"\n2)\t", rownames(ordered_area)[2])
```

```r
#determine the area of Scotland covered by water
cat("The area of Scotland covered by water is:\t",sum(scottish.lakes$area), "km^2")
```

## Exercise 2 - Crude Oil Production
The following CSV file https://drive.google.com/file/d/13WkinYhJNSr_XCldb3wpZOI0fgjer7Iw/view contains data
on crude oil prices from 1861 to 2020, measured in US dollars per barrel [2].

1) Write R code that is able to read the file and import it in a data frame structure. (Hint: before loading the file, open it with a text editor and check its structure).

2) Produce a plot with the Oil price as a function of the year

3) Which is the highest price in history ? When did it occur ?
 
3) Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative):

```r
#defining the df and changing the name of the fourth column for convenience
df <- as.data.frame(read.csv("crude-oil-prices.csv", header = TRUE ))
names(df)[4] <- "Price"

#check the output
cat("check the first rows:")
head(df,3)
```

2) Produce a plot with the Oil price as a function of the year

```r
options(repr.plot.width=6, repr.plot.height=5) 
plot (df$Year, 
      df$Price ,
      lwd=2, 
      col=" black ",
      xlab="Year", 
      ylab="Price per barrel [$]", 
      main="Price of crude oil in US $ per barrel vs Year", 
      pch=1,
      cex=0.1,
      cex.lab=0.9,
      cex.main=0.8,
      cex.axis=1
      )
lines ( df$Year , df$Price,col = "grey50" , lty = 1 )
```

3) Which is the highest price in history ? When did it occur ?

```r
#highest price determination from the df

cat("Highest price in history \t", max(df$Price))
cat("\nIt happened in\t \t\t", df$Year[which.max(df$Price)])
```

4) Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative):

```r
#evaluation of a vector of derivatives:
derivative <- df$Price[2:length(df$Price)] - df$Price[1:length(df$Price)-1]
#adding 0.5 to each year since the derivative is evaluated between two years
year       <- df$Year[-length(df$Year)] + 0.5

plot (year,
      derivative, 
      lwd=2, 
      col=" dark green ",
      xlab="Year", 
      ylab="Derivative of Price per barrel", 
      main="Derivative of the price of crude oil in US $ per barrel vs Year", 
      pch=1,
      cex=0.1,
      cex.lab=0.7,
      cex.main=0.9,
      cex.axis=0.7
      )
lines ( year  , derivative,col = "green" , lty = 1 )
```

## Exercise 3 - World Coal Production
• The following CSV file https://drive.google.com/file/d/1U1hK5o_d3vl8twwGaRJNcDJ79O9I5zJ1/view?usp=sharing
contains data on the coal production (in TW · hour) for several countries in the World [3].

1) Write R code that is able to read the file and import it in a tibble [4] structure

2) count the number of countries available in the file and produce a barplot with the number of entries for each country

3) selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions

4) for the 5 top Countries, create a plot of production as a function of time

5) generate a plot with the cumulative sum of the World’s coal production over the years

```r
library(tibble)
tbl <- as_tibble(read.csv("coal-production-by-country.csv", header = TRUE ))  #read as a tibble

names(tbl)[4] <- "Coal_prod"           #change name for convenienve
class(tbl)                               #check it is tibble

head(tbl,3)                            #verify structure of the file
```

2) count the number of countries available in the file and produce a barplot with the number of entries for each country

```r
#count the number of countries
cat("Number of different entities in the file:\t", length(unique(tbl$Entity)),"\n")

#actually we are counting also continents and organisations: we define a new tbl including only countries
library(dplyr)
tbl_countries = subset(tbl, !(Entity %in% c("Europe", "Africa", "North America","EU-28", "CIS", "Eurasia", "Europe (other)", 'Asia Pacific','Asia and Oceania', 'Central and South America', 'OECD','OPEC','Other Africa','Other Asia & Pacific', 'Other CIS','Other S. & Cent. America', 'USSR', "World")))
cat("Number of different countries in the file:\t", length(unique(tbl_countries$Entity)),"\n\n\n")
#levels(tbl$Entity)

#groupby the entity and count the occurences
cat("First 4 rows grouping by the Entity and counting the occurences:")
tbl_grouped <- tbl_countries %>% group_by(Entity) %>% count()
names(tbl_grouped) <- c('Entity', 'Counts')                    #set names to the columns
head(tbl_grouped,4)                                            #show the first 4 rows

options(repr.plot.width=12, repr.plot.height=4) 
#produce the barplot
barplot(tbl_grouped$Counts, 
        col = c("blue" ,"red" , "green" ,"yellow" , "orange"),
        lty="blank",
        xlab = "Countries",
        ylab = "Counts",
        main = "Total counts for each country",
        cex.lab=1.6
        )
```

3) selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions

```r
cat("selecting only the year after 1970:")
tbl_after1970 = tbl_countries[tbl_countries$Year>=1970, ]   #select data after or equal to 1970
head(tbl_after1970,3)                   #show first 3 rows

#summing after groupby entity:
cat("The total integrated production for each country is given by:")
tbl_total_prod <- tbl_after1970 %>% group_by(Entity) %>% summarise(total_prod = sum(Coal_prod))
head(tbl_total_prod,3)                  #show first 3 rows

cat("sorting in decreasing order of production, we find the top 5 countries:")
head(tbl_total_prod[order(tbl_total_prod$total_prod, decreasing = TRUE),],5)

```

4) For the 5 top Countries, create a plot of production as a function of time

```r
library(ggplot2)
options(repr.plot.width=10, repr.plot.height=5) 

#subset retaining only the rows corresponding to the top 5 countries
topc <- c("China", "United States", "Russia", "Australia", "India")
top5 <- subset(tbl_after1970, Entity %in% topc  ,select=c(Entity, Year, Code, Coal_prod))


ggplot(top5) + aes (x = Year, y = Coal_prod, group = Entity,         #plot grouping by entity
                    colour=Code) +                                   #different color for each entity(code)      
                    geom_line(size=1.3) +                            #linking with lines
                    labs(title = "Top 5 Producing Countries",
                        x = "Year",
                        y = "Coal production (TWh)",
                        colour = "Country")+
                        theme_linedraw()+                    
                        theme(text = element_text(size=10))     
```

5) generate a plot with the cumulative sum of the World’s coal production over the years

```r
#to have the cumulative we use the groupby (year) -> we neglect differences in Entity and sum over coal prod
tbl_cum_world <- tbl_after1970 %>% group_by(Year) %>% summarise(total_world_prod = sum(Coal_prod))#groupby year

#so far we have the annual production: the cumulative sum is needed:
tbl_cum_world$total_world_prod <- cumsum(tbl_cum_world$total_world_prod)                   #cumulative sum

cat("First 5 rows of the dataframe with cumulative production for each year:")
head(tbl_cum_world,5)

library(scales)
options(repr.plot.width=6, repr.plot.height=4) 
options(scipen=-2)

par(mgp=c(3,0.7,0))    #set distance label plot

plot (tbl_cum_world$Year, 
      tbl_cum_world$total_world_prod, 
      lwd=2, 
      col="red",
      xlab="Year", 
      ylab="Cumulative World Coal production (TWh)",
      main="Cumulative World Coal Production vs Year", 
      pch=1,
      las=1,        #horizontal labels in y axis
      cex=0.1,
      cex.lab=0.8,
      cex.main=0.7,
      cex.axis=0.6
      )


lines ( tbl_cum_world$Year, 
      tbl_cum_world$total_world_prod,col = "orange" , lty = 1 )

par(mgp=c(3,1,0))    #reset distances
```

As one can see, since there are no data available for Astrazeneca on the 01/11/2021 the method used so far fails.

The problem is related to data of Astrazeneca,since from the 1st November started arriving not on a daily basis.
This can be corrected summing the last available value of Astrazeneca. We need to complete the dataframe.

```r
library(tidyverse)

#extract the subset with astrazeneca
astra_subset <- subset(df_it, vaccine=='Oxford/AstraZeneca')

#complete with missing data
astra_subset$date<-as.Date(gsub(' SAST','',astra_subset$date))
astra_complete <- astra_subset %>%  
            complete(date = seq(from = as.Date(gsub(' SAST','',astra_subset$date[1])), 
                                to = as.Date(gsub(' SAST','',df_it$date[length(df_it$date)])),
                                by = '1 day'),
                     fill = list(vaccine='Oxford/AstraZeneca', location='Italy', total_vaccinations=NA))
astra_complete <- astra_complete %>% fill(total_vaccinations)

#write dates in usual way
astra_complete$date <- format(as.Date(astra_complete$date), "%Y-%m-%d")
```

Now that we have the dataframe complete we remove all the rows with AstraZeneca in the starting dataframe and we add the complete one,  finally we can replot the data.


```r
#removing rows with astra
df_it_noastra <- anti_join(df_it,astra_subset, by="vaccine")

#adding the rows with the complete astra dataframe (also when no changes occur on a daily basis):
df_it_merge <- rbind (df_it_noastra, astra_complete)
#ordering:
df_it_merge <- df_it_merge[order(df_it_merge$date),]
df_it_merge$date <- as.Date(gsub(' SAST','',df_it_merge$date))

# now try to replot:
df_corrected <- df_it_merge %>% group_by(date) %>% summarise(total_vaccinations=sum(total_vaccinations))
x1 <- as.Date(gsub(' SAST','',df_corrected$date[2:length(df_corrected$date)]))
y1 <- diff(df_corrected$total_vaccinations)

par(mgp=c(3,0.7,0))
plot (x1,
      y1, 
      lwd=2, 
      col="navy",
      xlab="Date", 
      ylab="Counts", 
      main="Number of vaccines shot per day in Italy", 
      pch=1,
      cex=0.4,
      cex.lab=0.7,
      cex.main=0.7,
      cex.axis=0.7,
      xaxt="n")
      #xlim=c(as.Date("2022-02-27"),as.Date("2022-03-10")))
      axis.Date(1, at=seq(min(x1), max(x1), by="1 month"), format="%d-%b")
      #ylim=c(0,10**6),
      
lines ( x1, y1,col = "blue" , lty = 1 )
par(mgp=c(3,1,0))
```

The same problem seems to occur, this time the same analysis tells us that the issue is related to the Johnson&Johnson when there are not changes within one day. We briefly repeat the same analysis:

```r
#extracting the subset:
jj_subset <- subset(df_it_merge, vaccine=='Johnson&Johnson')
jj_subset$date<-as.Date(gsub(' SAST','',jj_subset$date))
#filling with missing data
jj_complete <- jj_subset %>%  
            complete(date = seq(from = as.Date(gsub(' SAST','',jj_subset$date[1])), 
                                to = as.Date(gsub(' SAST','',df_it$date[length(df_it$date)])),
                                by = '1 day'),
                     fill = list(vaccine='Johnson&Johnson', location='Italy', total_vaccinations=NA))
jj_complete <- jj_complete %>% fill(total_vaccinations)
jj_complete$date <- format(as.Date(jj_complete$date), "%Y-%m-%d")

#removing all the johnson&johnson rows
df_it_nojj <- anti_join(df_it_merge,jj_subset, by="vaccine")
#adding the new rows:
df_it_merge <- rbind (df_it_nojj, jj_complete)
#ordering:
df_it_merge <- df_it_merge[order(df_it_merge$date),]
df_it_merge$date <- as.Date(gsub(' SAST','',df_it_merge$date))


#new df corrected
options(repr.plot.width=12, repr.plot.height=8) 
df_corrected <- df_it_merge %>% group_by(date) %>% summarise(total_vaccinations=sum(total_vaccinations))
x1 <- as.Date(gsub(' SAST','',df_corrected$date[2:length(df_corrected$date)]))
y1 <- diff(df_corrected$total_vaccinations)
plot (x1,
      y1, 
      lwd=2, 
      col="navy",
      xlab="Date", 
      ylab="Counts", 
      main="Number of vaccines shot per day in Italy", 
      pch=1,
      cex=0.4,
      cex.lab=1,
      cex.main=1,
      cex.axis=1,
      xaxt="n")
      #xlim=c(as.Date("2022-02-27"),as.Date("2022-03-10")))
      axis.Date(1, at=seq(min(x1), max(x1), by="1 month"), format="%b-%y")
      #ylim=c(0,10**6),
      
lines ( x1, y1,col = "blue2" , lty = 1 , lwd=2)
```

As a mater of fact, the number of total vaccinations by the manufacturer J&J decreases in that day and then comes back to the original value: this explains the huge fall followed by the peak. This is of course unphysical. We can fix it by changing the value to the average between the day before and the after, since we do not have any information for that day.

Doing this and replotting again, other smaller unphysical data appear for other manufacturers: we will fix all in once:

```r
#in case in one day there are less cumulative vaccines than the day before
#we average between day before and day after since this number must increase with time

manufacturers <- c("Johnson&Johnson", "Moderna","Pfizer/BioNTech")
for (i in 1:length(manufacturers)){
    a <- filter(df_us, vaccine==manufacturers[i])$total_vaccinations  #consider the one manufacturer at a time
    
    for (j in 2:(length(a)-1)){
    if (a[j] < a[j-1]) {                                              #is there is unphyiscal condition
        a[j] <- mean(c(a[j-1],a[j+1]))                                #do the average
        }
    }
    
   df_us[df_us$vaccine==manufacturers[i],]$total_vaccinations <- a
}



#redefine the dataframe grouped by date
df_us_tot <- df_us %>% group_by(date) %>% summarise(total_vaccinations=sum(total_vaccinations))

#now replot using the very same code as before
options(repr.plot.width=20, repr.plot.height=8) 
ggplot(df_us) + aes(    x= date, y = total_vaccinations,
                        group = vaccine, colour = vaccine)+ #different color for each manufacturer      
                geom_line(size=2) +                            #linking with lines
                labs(title = "Number of different vaccines given in USA vs Date",
                         x = "Date",
                         y = "Number of vaccines given",
                         colour = "Vaccine")+
                theme_linedraw()+
                theme(text = element_text(size=20),               #size of labels                
                        plot.title = element_text(hjust = 0.5),
                        axis.text.y = element_text(size=rel(1.5), hjust = 1),
                        axis.text.x = element_text(size=rel(1.5),hjust = 1,angle=60),
                        axis.title.y = element_text(size = rel(1.5), angle = 90),
                        axis.title.x = element_text(size = rel(1.5)),
                        legend.key.size = unit(1, 'cm'), #change legend key size
                        legend.key.height = unit(1, 'cm'), #change legend key height
                        legend.key.width = unit(3, 'cm'), #change legend key width
                        legend.title = element_text(size=24), #change legend title font size
                        legend.text = element_text(size=20))+

                scale_y_continuous(trans='log10')+

                scale_x_date(date_breaks = "3 months", date_labels="%m-%y")



options(repr.plot.width=10, repr.plot.height=6) 
y1 <- diff(df_us_tot$total_vaccinations)
plot (x1,
      y1, 
      lwd=2, 
      col="darkgreen",
      xlab="Date", 
      ylab="Counts", 
      main="Number of vaccines shot per day in USA", 
      pch=1,
      cex=0.2,
      cex.lab=1.5,
      cex.main=1.5,
      cex.axis=1.5,
      xaxt="n")
      axis.Date(1, at=seq(min(x1), max(x1), by="1 month"), format="%m-%y")
      lines ( x1, y1,col = "green" , lty = 1 )
```
